```python
# Plot the distribution of the 'origin' variable. How does the origin affect the other features in the dataset?
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset
file_path = r"C:\Users\praga\Downloads\mpg_raw.csv"
mpg_data = pd.read_csv(file_path)


# Plot the distribution of the 'origin' variable
plt.figure(figsize=(8, 6))
sns.countplot(x='origin', data=mpg_data)
plt.title('Distribution of Origin')
plt.xlabel('Origin')
plt.ylabel('Count')
plt.show()

# Analyze how the 'origin' affects other features
# Plotting distribution of 'mpg' by 'origin'
plt.figure(figsize=(10, 6))
sns.boxplot(x='origin', y='mpg', data=mpg_data)
plt.title('MPG by Origin')
plt.xlabel('Origin')
plt.ylabel('MPG')
plt.show()

# Plotting distribution of 'horsepower' by 'origin'
plt.figure(figsize=(10, 6))
sns.boxplot(x='origin', y='horsepower', data=mpg_data)
plt.title('Horsepower by Origin')
plt.xlabel('Origin')
plt.ylabel('Horsepower')
plt.show()
```
