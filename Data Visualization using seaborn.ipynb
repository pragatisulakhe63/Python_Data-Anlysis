## Relational Plots

```python
#line plot
sns.lineplot(data = df,x='sepal_length',y='sepal_width')
plt.show()
```

```python
# line plot
sns.lineplot(x='sepal_length',y='sepal_width',data=df,hue='species',ci=None)

plt.show()
```

## Scatter Plot

```python
sns.lineplot(data = df,x='sepal_width',y='sepal_length')
plt.title("Scatter Plot")
plt.show()
```

```python
sns.lineplot(data = df,x='sepal_width',y='sepal_length',hue = 'species')
plt.title("Scatter Plot")
plt.show()
```

## Red Plot

```python
sns.relplot(data =df,x='petal_length',y='sepal_width')
plt.show()
```

```python
sns.relplot(data =df,x='petal_length',y='sepal_width',kind='line',ci=None)
plt.show()
```

## Categorical plots

```python
##count plot value_count()
sns.countplot(x='species',data=df)
plt.show()
```

```python
sns.barplot(x='species',y='sepal_length',data=df)
plt.show()
```
