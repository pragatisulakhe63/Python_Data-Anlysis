```python
## Load the mpg-raw dataset and display the first 10 rows. what information is contained in this dataset?

import pandas as pd

# Load the dataset
file_path = r"C:\Users\praga\Downloads\mpg_raw.csv"
mpg_data = pd.read_csv(file_path)

# Display the first 10 rows
mpg_data.head(10)
```

```python
# Display the Data Types of all the variables from the above dataset 
mpg_data.dtypes
```

```python
# Display the duplicate rows from the above dataset using Python.
duplicate_rows = mpg_data[mpg_data.duplicated()]
print(duplicate_rows)
```

```python
# Display the categorical and continuous variables in the dataset using python.
categorical_vars = mpg_data.select_dtypes(include=['object']).columns
continuous_vars = mpg_data.select_dtypes(include=['float64', 'int64']).columns

print("Categorical Variables:")
print(categorical_vars)

print("\nContinuous Variables:")
print(continuous_vars)
```
